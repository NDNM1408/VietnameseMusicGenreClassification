```python
def extract_vocal_segment(audio_file, duration=60, threshold=0.01):
    y, sr = librosa.load(audio_file, sr=22050)
    vocal_start_time = 30
    # Calculate start frame and end frame based on vocal_start_time and duration
    start_frame = int(librosa.time_to_samples(vocal_start_time, sr=sr))
    end_frame = start_frame + int(sr * duration)

    # Ensure the end frame is within the audio length
    if end_frame > len(y):
        end_frame = len(y)

    # Extract the vocal segment
    vocal_segment = y[start_frame:end_frame]

    return vocal_segment, sr
```

```python
import os
path = '/kaggle/input/ballad-vietnamese/ballad'
list_audio = os.listdir(path)
```

```python
num_mfcc =20
sample_rate =22050
n_fft =2048
chroma_hop_length =512



data ={"filename": [], "chroma_stft_mean": [], "chroma_stft_var": [],
"rms_mean": [], "rms_var": [], "spectral_centroid_mean": [],
"spectral_centroid_var": [], "spectral_bandwidth_mean": [],
"spectral_bandwidth_var": [], "rolloff_mean": [],
"rolloff_var": [], "zero_crossing_rate_mean": [],
"zero_crossing_rate_var": [], "harmony_mean": [],
"harmony_var": [], "perceptr_mean": [],
"perceptr_var": [], "tempo" :[] ,"mfcc1_mean": [],
"mfcc1_var": [], "mfcc2_mean": [], "mfcc2_var": [],
"mfcc3_mean": [], "mfcc3_var": [], "mfcc4_mean": [],
"mfcc4_var": [], "mfcc5_mean": [] ,"mfcc5_var": [],
"mfcc6_mean": [], "mfcc6_var" :[] ,"mfcc7_mean": [],
"mfcc7_var": [], "mfcc8_mean" :[], "mfcc8_var": [],
"mfcc9_mean": [], "mfcc9_var": [], "mfcc10_mean": [],
"mfcc10_var": [], "mfcc11_mean": [], "mfcc11_var": [],
"mfcc12_mean": [], "mfcc12_var" :[], "mfcc13_mean" :[],
"mfcc13_var" : [],"mfcc14_mean": [] ,"mfcc14_var" :[],
"mfcc15_mean": [], "mfcc15_var": [], "mfcc16_mean": [],
"mfcc16_var": [], "mfcc17_mean": [], "mfcc17_var" : [],
"mfcc18_mean": [], "mfcc18_var" :[] , "mfcc19_mean": [],
"mfcc19_var": [], "mfcc20_mean" :[], "mfcc20_var" :[],
"label" :[]}
```

```python
import librosa
```

```python

for idx, file in enumerate(list_audio):
    audio_path = os.path.join(path, file)
    vocal_segment, sr = extract_vocal_segment(audio_path)
    # Chromagram
    chromagram = librosa.feature.chroma_stft(y=vocal_segment, sr=sr, hop_length=chroma_hop_length)
    data["chroma_stft_mean"].append(chromagram.mean())
    data["chroma_stft_var"].append(chromagram.var())
    
    RMSEn = librosa.feature.rms(y=vocal_segment)
    data["rms_mean"].append(RMSEn.mean())
    data["rms_var"].append(RMSEn.var())
    
    # Spectral Centroid
    spec_cent = librosa.feature.spectral_centroid(y=vocal_segment)
    data["spectral_centroid_mean"].append(spec_cent.mean())
    data["spectral_centroid_var"].append(spec_cent.var())
    
    # Spectral Bandwidth
    spec_band = librosa.feature.spectral_bandwidth(y=vocal_segment, sr=sr)
    data["spectral_bandwidth_mean"].append(spec_band.mean())
    data["spectral_bandwidth_var"].append(spec_band.var())
    
    # Rolloff
    spec_roll = librosa.feature.spectral_rolloff(y=vocal_segment, sr=sr)
    data["rolloff_mean"].append(spec_roll.mean())
    data["rolloff_var"].append(spec_roll.var())
    
    # Zero Crossing Rate
    zero_crossing = librosa.feature.zero_crossing_rate(y=vocal_segment)
    data["zero_crossing_rate_mean"].append(zero_crossing.mean())
    data["zero_crossing_rate_var"].append(zero_crossing.var())
    
    # Harmonics and Percussive
    harmony, perceptr = librosa.effects.hpss(y=vocal_segment)
    data["harmony_mean"].append(harmony.mean())
    data["harmony_var"].append(harmony.var())
    data["perceptr_mean"].append(perceptr.mean())
    data["perceptr_var"].append(perceptr.var())
    
    tempo, _ = librosa.beat.beat_track(y=vocal_segment, sr=sr)
    data["tempo"].append(tempo)
    
    mfcc = librosa.feature.mfcc(y=vocal_segment, sr=sr, n_mfcc=num_mfcc, n_fft=n_fft, hop_length=chroma_hop_length)
    mfcc = mfcc.T
    for x in range(20):
        feat1 = "mfcc" + str(x+1) + "_mean"
        feat2 = "mfcc" + str(x+1) + "_var"
        data[feat1].append(mfcc[:, x].mean())
        data[feat2].append(mfcc[:, x].var())
    print(idx)
```

```python
data["label"] = []
data["filename"] = []
for file in list_audio:
    data["label"].append("ballad")
    data["filename"].append(file[:-4])
```

```python
import pandas as pd
df = pd.DataFrame(data)
```
